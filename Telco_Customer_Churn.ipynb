# Handling numerical and non-numerical data separately

```python
# separate numerical and non numerical data
numerical_df = df.select_dtypes(include=['int64', 'float64'])
categorical_df = df.select_dtypes(include=['object', 'category'])
```

```python
# process numerical data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_df_scaled = scaler.fit_transform(numerical_df)
```

```python
# process non numerical data
numerical_df_encoded = pd.get_dummies(categorical_df)
```

```python
# combine the preprocessed numerical and non numerical data
import numpy as np

x_combined = np.hstack((numerical_df_scaled, numerical_df_encoded))
```

```python
# model training
clf.fit(x_combined, y_train)
```
